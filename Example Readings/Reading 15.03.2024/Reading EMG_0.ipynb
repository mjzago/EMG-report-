```python
# Store evething on a variable files (CHANGE FILENAME IF NEEDED)

files = [
    'data-F_on_body_with_movements_and_acc.csv',
    'data-F_on_body_with_movements.csv',
    'data-F_on_body.csv',
    'data-F_plastic.csv'
]
```

```python
import pandas as pd

# List to store the emg_0 DataFrames from each file
emg_0_dataframes = []

# Read and parse only the emg_0 column from each CSV file
for filename in files:
    # Read the CSV file
    df = pd.read_csv(filename)
    
    # Add the emg_0 column to the DataFrame
    emg_0_dataframes.append(df['emg_0'])

# Concatenate all emg_0 DataFrames into a single DataFrame
emg_0_combined = pd.concat(emg_0_dataframes)

# Perform statistical analysis on the emg_0 column
statistics = emg_0_combined.describe()

print(statistics)
```

```python
import pandas as pd

# Function to calculate statistics of the emg_0 column
def calculate_statistics(df):
    return df['emg_0'].describe()

# Read and analyze only the emg_0 column from each CSV file
for filename in files:
    # Read the CSV file
    df = pd.read_csv(filename)
    
    # Calculate statistics for the emg_0 column
    statistics = calculate_statistics(df)
    
    # Print statistics for the current file
    print(f"Statistics for the file {filename}:")
    print(statistics)
    print('\n')  # Add a blank line between results of different files

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read and plot histogram for the emg_0 column of each CSV file
for filename in files:
    # Read the CSV file
    df = pd.read_csv(filename)
    
    # Plot histogram for the emg_0 column
    plt.figure(figsize=(8, 6))
    plt.hist(df['emg_0'], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of the emg_0 column - {filename}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```
